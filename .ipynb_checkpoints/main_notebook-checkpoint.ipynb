### 3- NBA

```python
import pandas as pd
import requests

res = requests.get('https://www.balldontlie.io/api/v1/players')
```

```python
res
```

```python
data = res.json()
data
```

```python
df = pd.DataFrame(data['data'])
df
```

```python
df.head(4)
```

```python
df['team']
```

```python
def flatten(df, col_list):
    for column in col_list:
        flattened = pd.DataFrame(dict(df[column])).transpose()
        columns = [str(col) for col in flattened.columns]
        flattened.columns = [column + '_' + colname for colname in columns]
        data = pd.concat([df, flattened], axis=1)
        data = data.drop(column, axis=1)
    return data
```

```python
df
```

```python
nested_columns = ['team']
```

```python
flat = flatten(df, nested_columns)
flat.head(50)
```

A possible project is creating a basketball players database and try to find a correlation between their phsyical stats, position, and other variables to their current market value, so we can build a recommender on which players to hire
